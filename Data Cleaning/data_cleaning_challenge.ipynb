## Cleaning Null Values

Let us drop the Nan row in the 1st index.

```python
raw_data_drop_na = raw_data[~raw_data["Row Type"].isnull()]
raw_data_drop_na
```

Let us now get rid of the last two columns which are just null values

```python
raw_data_drop_col = raw_data_drop_na.drop(columns = ['Unnamed: 9', 'Unnamed: 10'])
raw_data_drop_col
```

The main idea here is to wap up all of this data into a long format instead of the current wide format so that we have first and last name as columns.
