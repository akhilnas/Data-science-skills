# Data Cleaning Challenge 1

```python
## Required Libraries
import pandas as pd
import numpy as np
```

```python
# Read in Data
raw_data = pd.read_csv('data_cleaning_challenge.csv')
raw_data
```

## Cleaning Null Values

Let us drop the Nan row in the 1st index.

```python
raw_data_drop_na = raw_data[~raw_data["Row Type"].isnull()]
raw_data_drop_na
```

Let us now get rid of the last two columns which are just null values

```python
raw_data_drop_col = raw_data_drop_na.drop(columns = ['Unnamed: 9', 'Unnamed: 10'])
raw_data_drop_col
```

The main idea here is to wap up all of this data into a long format instead of the current wide format so that we have first and last name as columns.

## Data Transformation

```python
Name_iterations = []
counter = 0

for i in raw_data_drop_col['Row Type']:
    if 'first name' in i:
        counter += 1
    Name_iterations.append(counter)
Name_iterations
```

```python
iter_cols = raw_data_drop_col.copy()
iter_cols["Iteration"] = Name_iterations
iter_cols
```

Let us Get Rid of Rows that are pointless headers such as Row Type, Iter Number etc.

```python
data_rows_removed = iter_cols[iter_cols['Row Type'] != 'Row Type']
data_rows_removed
```

Data Cleaning the Names Dataframe

```python
# Acquire all Names from Table
names = data_rows_removed[data_rows_removed['Row Type'].str.contains('first name')]
names = names.drop(columns = ['Speed1', 'Speed2', 'Electricity', 'Effort', 'Weight', 'Torque'])
names.rename(columns = {"Row Type": "First Name", "Iter Number": "Last Name", "Power1": "Date"}, inplace=True)
names
```

Let us clean up this Names Dataframe as we do not want the text "First Name', 'Last Name' etc. in the table. Also Let us use Proper Capitalization for the words.

```python
names["First Name"] = names["First Name"].str[12:]
names["Last Name"] = names["Last Name"].str[11:]
names["Date"] = names["Date"].str[6:]

# Capitalizing
names["First Name"] = names["First Name"].str.title()
names["Last Name"] = names["Last Name"].str.title()
names
```

```python
no_names_dataframe = data_rows_removed[~data_rows_removed['Row Type'].str.contains('first name')]
no_names_dataframe
```

Let us now merge these two previous Dataframes based on Iteration Number

```python
clean_table = pd.merge(left = names, right = no_names_dataframe, how="inner", on = "Iteration")
clean_table = clean_table.drop(columns=['Iteration'])
clean_table
```

The cleaned Data is presented above. There is an alternative method to doing this operation.

```python
# Original Partial Cleaned Data
raw_data_drop_col
```

```python
cleaning_alternative_dataframe = raw_data_drop_col
cleaning_alternative_dataframe['First Name'] = cleaning_alternative_dataframe.apply(lambda x: x['Row Type'] if 'first name' in x['Row Type'] else np.nan, axis=1)
cleaning_alternative_dataframe['Last Name'] = cleaning_alternative_dataframe.apply(lambda x: x['Iter Number'] if 'first name' in x['Row Type'] else np.nan, axis=1)
cleaning_alternative_dataframe['Date'] = cleaning_alternative_dataframe.apply(lambda x: x['Power1'] if 'first name' in x['Row Type'] else np.nan, axis=1)
cleaning_alternative_dataframe
```

Now we can fill the columns up using the Pandas fillna function with the method ffill as these will ensure a forward filling procedure.

```python
cleaning_alternative_dataframe["First Name"].fillna(method='ffill', inplace=True)
cleaning_alternative_dataframe["Last Name"].fillna(method='ffill', inplace=True)
cleaning_alternative_dataframe["Date"].fillna(method='ffill', inplace=True)
cleaning_alternative_dataframe
```

Now let us drop the unnecessary rows where the Row Type Column contains the names.

```python
cleaning_alternative_dataframe_rows = cleaning_alternative_dataframe[~cleaning_alternative_dataframe["Row Type"].str.contains('first name')]
cleaning_alternative_dataframe_rows
```

Let us also drop the unnecessary header rows

```python
cleaning_alternative_dataframe_rows = cleaning_alternative_dataframe_rows[cleaning_alternative_dataframe_rows["Row Type"] != "Row Type"]
cleaning_alternative_dataframe_rows
```

Let us now rearrange the colums so that the names and date are the first columns.

```python
cols = ['First Name', 'Last Name', 'Date', 'Row Type', 'Iter Number', 'Power1', 'Speed1', 'Speed2', 'Electricity', 'Effort', 'Weight', 'Torque']
alternate_dataframe = cleaning_alternative_dataframe_rows[cols]
alternate_dataframe
```

```python
alternate_dataframe["First Name"] = alternate_dataframe["First Name"].str[12:]
alternate_dataframe["Last Name"] = alternate_dataframe["Last Name"].str[11:]
alternate_dataframe["Date"] = alternate_dataframe["Date"].str[6:]

# Capitalizing
alternate_dataframe["First Name"] = alternate_dataframe["First Name"].str.title()
alternate_dataframe["Last Name"] = alternate_dataframe["Last Name"].str.title()
alternate_dataframe
```

Thus the data is cleaned and now we can save this cleaned data to a csv file.

```python
# Saving Dataframe
alternate_dataframe.to_csv('clean_data_challenge_1.csv', index=False)
```
